```python
# append() : 맨 뒤에 1개 추가
fruits = ["apple", "banana"]
fruits.append("orange")
print(fruits)
# insert() : 원하는 위치에 추가

fruits = ["apple", "banana"]
fruits.insert(1, "orange")
print(fruits)
# extend() : 여러 개 추가

nums = [1, 2, 3]
nums.extend([4, 5, 6])
print(nums)
# remove() : 값으로 삭제

fruits = ["apple", "banana", "orange"]
fruits.remove("banana")
print(fruits)

# 수정은 인덱스 할당, 슬라이싱 할당으로 수정
# pop() : 인덱스로 삭제 후 반환
nums = [10, 20, 30]
result = nums.pop() # 마지막 인덱스 삭제 후 반환
print(result)
print(nums)

nums = [10, 20, 30]
result = nums.pop(1) # 인덱스1 삭제 후 반환
print(result)
print(nums)
# clear() : 전체 삭제
nums = [10, 20, 30]
nums.clear()
print(nums)
```
