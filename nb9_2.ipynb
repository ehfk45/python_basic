```python
def calc(a, b):
    x = a + b
    y = a - b
    z = a * b
    return x , y, z # 3개의 값을 반환하는 것이 아니라 tuple로 묶어서 반환


r1 = calc(1,2)
print(r1)

r1,r2,r3 = calc(10,5) # 언패킹
print(r1)
print(r2)
print(r3)
```
