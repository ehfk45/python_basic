```python
def hello():
    print("안녕하세요!")

# 함수를 변수에 저장
f = hello

# 변수로 함수 호출
f()
```

```python
def hello():
    print("안녕하세요!")

def execute(func):
    func() # 함수호출연사자를 사용했다는 것은 func매개변수값이 함수다

execute(hello)
```

```python
# 일반 함수 정의
def add1(a, b):
    return a + b

# 람다 함수
add2 = lambda a, b : a + b # add2에는 "lambda a, b : a + b" 람다가 반환한 함수(객체)가 대입

r1 = add1(5, 1)

print(r1)

r2 = add2(1, 3)
print(r2)
```

```python
# 람다를 익명함수라고 부르기도 한다


r3 = (lambda a, b : a + b)(10,50)
print(r3)


```

```python
def score(item): # 아이템을 입력받아 1번째 요소를 반환 ex)  ('홍길동', 90)가 인수로 입력되면 90 반환
    return item[1]

s = [
    ('홍길동', 90),
    ('이순신', 85),
    ('강감찬', 100)
]

r = sorted(s, key=score, reverse=True)

print(r)

r2 = sorted(s, key=lambda e: e[1], reverse=False) # 람다를 이용하여 score함수 대신 사용
print(r2)
```
