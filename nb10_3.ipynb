```python
import datetime
import time
```

```python
now = datetime.datetime.now()
print(now)
```

```python
print(now.year)
print(now.month)
print(now.day)
```

```python
today = datetime.date.today()
print(today)
```

```python
print(today.weekday())

```

```python
weekday = today.weekday()

week = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일'] #0:월

print(week[weekday])

xmas = datetime.date(2026, 12 , 25)
print(week[xmas.weekday()])
```

```python
print(xmas - today) # 크리스마스까지 남은 일수

print(today + datetime.timedelta(days=1000)) # 1000일 후 날짜

print(today - datetime.date(2026,6,19))
```
