```python
# with문 사용하지 않는 경우

try:    
    f = open("students.csv", "r", encoding="utf-8")
    print(f.read())
    print(10 / 0)  # ZeroDivisionError 발생
finally:    
    f.close()
    print('자원해지 성공')
```

```python
# with문을 사용하는 경우
with open("students.csv", "r", encoding="utf-8") as f:
    print(f.read())
    print(10 / 0)  # ZeroDivisionError 발생
```

```python
names = []
scores = []


with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        # print(name, int(score))
        names.append(name)
        scores.append(score)


print(names)
print(scores)   
```

```python
students = {}

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        name, score = line.strip().split(",")
        students[name] = int(score)

print(students)
```

```python
s = []

with open("students.txt", "r", encoding="utf-8") as f:
    for line in f:
        a = line.strip().split(',')

        s.append(tuple(a))

print(s)        
```
