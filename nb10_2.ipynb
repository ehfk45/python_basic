```python
import random
# random.random() 함수만으로 로또

# 공을 45 -> 변수 45

balls = list(range(1,46))
print(balls)
```

```python
# balls -> shuffle() -> 섞기
for i in range(0, 1000000):
    temp = balls[0]
    r_num = int(random.random() * len(balls)) # 0.000 ~ 44.999
    #r_num = math.floor(random.random() * len(balls)) # 0.000 ~ 44.999
    balls[0] = balls[r_num]
    balls[r_num] = temp

print(balls)
print(balls[:6])

```

```python
# random 모듈안 sample()함수를 사용
lotto = random.sample(range(1,46),6)
print(lotto)
lotto.sort()
print(lotto)


```

```python
for i in range(10):
    lotto = random.sample(range(1,46),6)
    lotto.sort()
    print(f"{i}번쨰 로또 번호: {lotto}")
        
```

```python
def my_lotto_create(a):
    result = []
    for i in range(a):
        lotto = random.sample(range(1,46),6)
        lotto.sort()
        #print(f"{i+1}번쨰 로또 번호: {lotto}")
        result.append(lotto)
    return tuple(result)
```

```python
result = my_lotto_create(3)
print(result)
```

```python
def dice_cast():
    d1 = random.randint(1,6)
    d2 = random.randint(1,6)
    return d1 + d2

#person1_dice = {'name': '홍길동', 'dice_value': 0}
#person1_dice['dice_value'] = dice_cast()
person1_dice = dice_cast()
person2_dice = dice_cast()

if person1_dice == person2_dice:
    print(f'비겼다. person1_dice: {person1_dice}, person2_dice: {person2_dice}')
elif person1_dice > person2_dice:
    print(f'person1_dice Win. person1_dice: {person1_dice}, person2_dice: {person2_dice}')
else:
    print(f'person2_dice Win. person1_dice: {person1_dice}, person2_dice: {person2_dice}')        
```
