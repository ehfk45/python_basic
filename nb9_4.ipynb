```python
nums = [1, 2, 3, 4]

# 리스트의 각 요소(x)에 2를 곱한 결과를 반환
result = map(lambda x: x * 2, nums)

print(result) #[2, 4, 6, 8]

print(list(result))
```

```python
nums = [1, 2, 3, 4, 5]

# 짝수인 요소만 선택
result = list(filter(lambda x: x % 2 == 0, nums))

print(result)
```

```python
names = ["banana", "apple", "kiwi"]

# 문자열의 길이를 기준으로 오름차순 정렬
result = sorted(names, key=lambda x: len(x), reverse=True)

print(result)
```

```python
students = [
    {"이름": "홍길동", "나이": 20, "점수": 90},
    {"이름": "이순신", "나이": 18, "점수": 85},
    {"이름": "김유신", "나이": 22, "점수": 95}
]

# "나이" 값을 기준으로 오름차순 정렬
result = sorted(students, key=lambda x: x["나이"])
print(result)

# x : 리스트의 각 딕셔너리
# x["나이"] : 각 딕셔너리의 "나이" 값을 꺼냄
# sorted()는 이 값을 기준으로 정렬한다.
```

```python
students = [
    {"이름": "홍길동", "반": 2, "점수": 90},
    {"이름": "이순신", "반": 1, "점수": 90},
    {"이름": "김유신", "반": 1, "점수": 95}
]

# 1순위: 반, 2순위: 점수
result = sorted(students, key=lambda x: (x["반"], x["점수"]))
print(result)
```

```python
names = ["banana", "apple", "kiwi"]

# 문자열 길이가 가장 짧은 요소 반환
result = min(names, key=lambda x: len(x))

print(result) # kiwi
```

```python
nums = [1, 3, 5, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 하나라도 True이면 True 반환

r1 = map(lambda x: x % 2 == 0, nums)
print('r1:',r1)
r2 = list(r1)
print('r2:',r2)


result = any(map(lambda x: x % 2 == 0, nums))

print(result) # True
```

```python
nums = [2, 4, 6, 8]

# 각 요소가 짝수인지(True/False)로 변환
# 모두 True이면 True 반환
result = all(map(lambda x: x % 2 == 0, nums))

print(result) #True
```
