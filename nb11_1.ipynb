```python
import csv 
```

```python
# 파일같은 자원 -> 메모리 로드 후 자원해지
with open('s.csv','r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    print(type(reader))
    data = list(reader)

data    
```

```python
# 변수 설정
names = []
kors = []
engs = []

names = [r['이름'] for r in data]
kors = [int(r['국어']) for r in data]
engs = [int(r['영어']) for r in data]
print(names)
print(kors)
print(engs)
```

```python
name_kor = list(zip(names, kors))
name_eng = list(zip(names, engs))
print(name_kor)
print(name_eng)

name_kor_list = list(zip(names, kors)) # 반복자 커스 이동 처음 -> 끝
name_eng_list = list(zip(names, engs))
name_kor_dict = dict(zip(names, kors))
name_eng_dict = dict(zip(names, engs))

print(name_kor_list)
print(name_eng_list)
print(name_kor_dict)
print(name_kor_dict)


#name_kor_eng = zip(names, kors, engs)
print(list(name_kor_eng))
name_avg_list = [(n, (k + e) / 2) for n, k, e in zip(names, kors, engs)]
print(name_avg_list)
```
