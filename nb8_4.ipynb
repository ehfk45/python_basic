```python
import csv

# open()은 내장함수
with open("students.csv", "r", encoding="utf-8") as f:
    # reader() : csv모듈안에 있는 함수
    reader = csv.reader(f) # <class '_csv.reader'>
    print(type(reader))

    for row in reader:
        print(row)
        print(type(row)) # 리스트 클래스
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    next(reader)

    for row in reader:
        print(row)
        
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f) # 헤드라인은 키로, 한줄의 내용은 값으로 

    for row in reader:
        print(row)
```

```python
with open("students.csv", "r", encoding="utf-8") as f:
    reader = csv.DictReader(f)

    data = list(reader)

print(data)

data[2]['영어'] = 80

print(data[2])
```

```python
with open("students_result.csv", "w", newline="", encoding="utf-8-sig") as f:
    # data[0].keys() → 첫 번째 딕셔너리의 키(컬럼명) 사용    
    writer = csv.DictWriter(f, fieldnames=data[0].keys())
    # 헤더(컬럼명) 저장
    writer.writeheader()   
    # data 리스트에 있는 모든 딕셔너리를 한 번에 저장
    writer.writerows(data)  

print("CSV 저장 완료")
```
