```python
try:
    num = int(input("숫자 입력 : "))
    result = 10 / num
    print(result)

    print('자원해지')

except ValueError:
    print("숫자를 입력하세요.")

except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")


print('종료')     
```
