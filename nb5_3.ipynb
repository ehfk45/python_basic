```python
nums = [10, 20, 30, 40]

print(nums[0])
print(nums[-1])

# 인덱스 할당 연산
nums[2] = 99 # 리스트는 가변 객체
print(nums)

n2 = (3,5)
print(n2[0])
n2[0] = 999 #오류가 뜸
```

```python
nums = [10, 20, 30, 40, 50]

print(nums[1:4]) # [20, 30, 40]
print(nums[:3]) # [10, 20, 30]
print(nums[2:]) # [30, 40, 50]
print(nums) #  [10, 20, 30, 40, 50] 슬라이싱 연산 후에도 리스트 자신은 변경되지 않는다. 

# 역순 출력
nums = [10, 20, 30, 40, 50]
print(nums[::-1]) # [50, 40, 30, 20, 10]

print(nums)
```

```python
nums = [50, 10, 40, 20]
result = sorted(nums)
print(nums)
print(result)
```

```python
# 정렬 메서드
nums = [50, 10, 40, 20]
nums.sort() # 오름차순
print(nums)

nums = [50, 10, 40, 20]
nums.sort(reverse=True) # 내림차순
print(nums)
```
