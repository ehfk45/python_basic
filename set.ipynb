```python
a = {1, 2, 3}
b = {3, 4, 5}

# 수학용 집합 연산이 가능하다
print(a & b)
print(a | b)
print(a - b)
```

```python
a = {1, 2, 3, 2}
print(a) # 중복된 값은 제거된다

print(list(a)[0]) # 인덱싱은 불가하지만 리스트로 변환하면 가능하다
```

```python
a = {"1":"admin","pw":1234}
print(a["1"]) # 'admin' 출력

b = {3,4,6}
print(b,type(b)) # set <class 'set'>

c=[]
print(c,type(c)) # list <class 'list'>

d = {}
d2 = dict()
print(d,type(d)) # dict <class 'dict'>
```

```python
a = {1, 2, 3}
b = {3, 4, 5}

# 교집합
print(a & b)
print(a.intersection(b))

# 합집합
print(a | b)
print(a.union(b))   

# 차집합
print(a - b)    
print(a.difference(b))

```
