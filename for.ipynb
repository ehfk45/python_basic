```python
fruits = ["apple", "banana", "orange"]

for fruit in fruits:
    print(fruit)
```

```python
student = {
    "name": "홍길동",
    "age": 20
}
# 키
for key in student:
    print(key)
# 값
for value in student.values():
    print(value)

# 키 + 값 여기서 언패킹이 발생함.
for key, value in student.items():
    print(key, value)
```
