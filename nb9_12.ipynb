```python
def hello():
    print("Hello")

f = hello      # 함수 자체를 저장 (괄호 없음)

f()
```

```python
def square(x):
    return x * x

def test(fun): # 고차 함수
    return fun(5)

print(test(square))
```

```python
def outer():
    def inner():
        print("inner")

    return inner

f = outer()

f()

outer()()
```

```python
def plus(x):
    return x + 10

def minus(x):
    return x - 10

funcs = [plus, minus]

print(funcs[0](100))
print(funcs[1](100))
```

```python
def add(x, y):
    return x + y

def sub(x, y):
    return x - y

calc = {
    "+": add,
    "-": sub
}

print(calc["+"](10, 5))
print(calc["-"](10, 5))
```
