```python
import csv

file = open ("score.csv", "r", encoding="utf-8")
reader = csv.DictReader(file)

data = list(reader)
file.close()
```

```python
print(data)
```

```python
for row in data:

    print(row)
```

```python
for row in data:

    if row['나이'] =='':
        print(row['이름'],'나이 결측')
        
    if row['국어'] =='' or row['국어'] == '결측':
        print(row['이름'],'국어 결측')

    if row['영어'] =='' or row['영어'] == 'NaN':
        print(row['이름'], '영어 결측')
                 
```

```python
for row in data:

    if row["나이"] != "":
        age = int(row["나이"])

        if age < 0 or age > 100:
            print(row["이름"], "나이 이상치")
            
```

```python
for row in data:

    if row["국어"] not in ["", "결측"]:

        korean = int(row["국어"])

        if korean < 0 or korean > 100:
            print(row["이름"], "국어 이상치")
```

```python
for row in data:

    if row["영어"] not in ["", "NaN"]:

        english = int(row["영어"])

        if english < 0 or english > 100:
            print(row["이름"], "영어 이상치")
```

```python
for row in data:

    try:
        kor = int(row["국어"])
        eng = int(row["영어"])

        if 0 <= kor <= 100 and 0 <= eng <= 100:
            avg = (kor + eng) / 2
            row["평균"] = avg

    except:
        continue
```

```python
for row in data:
    print(row)
```

```python
for row in data:

    if "평균" in row:

        if row["평균"] >= 90:
            row["합격"] = "합격"

        else:
            row["합격"] = "불합격"
```

```python
for row in data:
    print(row)
```

```python
avg_list = []

for row in data:

    if "평균" in row:
        avg_list.append(row["평균"])
```

```python
mean = sum(avg_list) / len(avg_list)

print("평균 :", mean)

print("최대 :", max(avg_list))

print("최소 :", min(avg_list))
```

```python
avg_list.sort()

n = len(avg_list)

median = avg_list[n//2]

print("중앙값 :", median)
```
