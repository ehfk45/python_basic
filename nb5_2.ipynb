```python
a = np.array([[1, 2],
              [3, 4]])
print(a)

b = np.array([[5, 6],
              [7, 8]])
print(a + b) 
# print(a * b) 곱하기는 안씀 
# [1+5  2+6] 
# [3+7  4+8]
```

```python
import numpy as np

a = np.array([[1, 2, 3],
              [4, 5, 6]])   # 2행3열

b = np.array([[10],
              [20],
              [30]])        # 3행1열

print(a @ b)      #결과는 2행1열  
```
