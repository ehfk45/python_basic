```python
scores = [95, 60, 80, 45]

for score in scores:
    if score >= 70:
        print("합격")
    else:
        print("불합격")


numbers = [3, 8, 11, 20, 25]

for num in numbers:
    if num % 2 == 0:
        print(f"{num} : 짝수")
    else:
        print(f"{num} : 홀수")


fruits = ["apple", "banana", "kiwi", "orange"]

for fruit in fruits:
    if len(fruit) >= 6:
        print(fruit)
```
