```python
scores = []

for order in orders:
    if order["구매금액"] is not None:  # 구매금액이 결측치가 아닌 경우에만 추가
        scores.append(order["구매금액"])    
print(scores)  # 구매금액 리스트 출력

avg = sum(scores) / len(scores)  # 평균 계산
print(avg)  # 평균 출력

for order in orders:
    if order["구매금액"] is None:  # 구매금액이 결측치인 경우
        order["구매금액"] = avg  # 평균값으로 대체
```
